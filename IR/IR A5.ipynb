```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
# Define the starting URL
start_url = "https://example.com"  # Replace with the website you want to crawl
max_pages = 10  # Maximum number of pages to crawl
```

```python
# Initialize a list to store visited URLs
visited_urls = []
```

```python
# Initialize a list to store URLs to visit
to_visit = [start_url]
```

```python
# Crawl the web
while to_visit and len(visited_urls) < max_pages:
    # Get the next URL to visit
    current_url = to_visit.pop(0)
    
    # Skip this URL if it has already been visited
    if current_url in visited_urls:
        continue
    
    try:
        # Send an HTTP GET request to the current URL
        response = requests.get(current_url)
        
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content of the page
            soup = BeautifulSoup(response.text, 'html.parser')
            
            # Extract and print the page title
            page_title = soup.title.text if soup.title else "No Title"
            print(f"Title: {page_title}")
            
            # Extract and print all the links on the page
            for link in soup.find_all('a'):
                href = link.get('href')
                if href:
                    full_url = urljoin(current_url, href)
                    print(f"Link: {full_url}")
                    to_visit.append(full_url)
            
            # Mark the current URL as visited
            visited_urls.append(current_url)
        else:
            print(f"Failed to fetch URL: {current_url}")
    
    except Exception as e:
        print(f"Error processing URL: {current_url}\n{e}")

print("Web crawling completed.")
```
